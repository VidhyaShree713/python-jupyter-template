### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for different plots
x = np.linspace(0, 10, 50)
y = np.sin(x)

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart (Top-Left)
axes[0, 0].plot(x, y, color='blue', linestyle='-', linewidth=2)
axes[0, 0].set_title("Line Chart")

# Bar Chart (Top-Right)
axes[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
axes[0, 1].set_title("Bar Chart")

# Scatter Plot (Bottom-Left)
axes[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o', s=100)
axes[1, 0].set_title("Scatter Plot")

# Pie Chart (Bottom-Right)
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'], startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
